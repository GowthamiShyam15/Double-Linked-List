# Double LL

```python
#Creating Node

class Node:
    def __init__(self,data):
        self.data=data
        self.prev=None
        self.next=None

#Creating Link Between Nodes

class DLL:
    def __init__(self):
        self.head=None
        
    #insert at begining

    def insert_begin(self,data):
        nb=Node(data)
        temp=self.head
        temp.prev=nb
        nb.next=temp
        self.head=nb
        
    #insert at end
    
    def insert_end(self,data):
        ne=Node(data)
        temp=self.head
        while temp.next is not None:
            temp=temp.next
        temp.next=ne
        ne.prev=temp  
        
    #insert at specified position

    def insert_sp(self,pos,data):
        np=Node(data)
        temp=self.head
        for i in range(pos-1):
            temp=temp.next
        np.prev=temp
        np.next=temp.next
        temp.next.prev=np
        temp.next=np
        
    #Delete at begining

    def delete_begin(self):
        temp=self.head
        self.head=temp.next
        temp.next=None
        self.head.prev=None
        
    #Delete at end

    def delete_end(self):
        before=self.head
        temp=self.head.next
        while temp.next is not None:
            temp=temp.next
            before=before.next
        before.next=None
        temp.prev=None
        
    #Delete at specified position

    def delete_sp(self,pos):
        before=self.head
        temp=self.head.next
        for i in range(pos-1):
            temp=temp.next
            before=before.next
        before.next=temp.next
        temp.next.prev=before
        temp.prev=None
        temp.next=None
        
    #searching
    def search(self):
        x = int(input("\nEnter an element to be serched in LL: "))
        temp = self.head
        pos = 0
        flag = False
        
        #checks whether list is empty
        if self.head == None:
            print("Empty")
        else:
            while temp != None:
            
            #Compares node to be found with each node present in list
                if x == temp.data:
                    flag = True
                    break
                pos += 1    
                temp = temp.next
    
            if flag == True:
                print(x," is found in the list at position ",pos)
            else:
                print(x," is not found")
                
        
#To display
    def display(self):
        if self.head is None:
            print("EMPTY!")
        else:
            temp=self.head
        while temp:
            print(temp.data,"<--> ",end="")
            temp=temp.next
    
          
#To access Nodes create object
L=DLL()
n1=Node(10)
L.head = n1
n2 = Node(20)
n1.next = n2
n2.prev = n1
n3 = Node(30)
n2.next = n3
n3.prev = n2

print("Creation of Nodes:")
L.display()
print()

print("\nAfter Inserting Node at Beginning :")
L.insert_begin(5)
L.display()

print()

print("\nAfter Inserting Node at Ending :")
L.insert_end(40)
L.display()

print()

print("\nAfter Inserting Node at specific position :")
L.insert_sp(3,100)
L.display()

print()

print("\nAfter Deleting Node at Beginning :")
L.delete_begin()
L.display()

print()

print("\nAfter Deleting Node at Ending :")
L.delete_end()
L.display()

print()

print("\nAfter Deleting Node at specific position :")
L.delete_sp(2)
L.display()

print()

L.search()
```
